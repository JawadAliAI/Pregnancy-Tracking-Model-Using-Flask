```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder instance for binary columns
label_encoder = LabelEncoder()

# 1. Label Encoding for binary categorical features
binary_columns = ['Fetal Movement', 'Contractions', 'Swelling', 'Smoking', 'Alcohol Consumption','Symptoms',
                  'Dietary Habits', 'Mood Changes', 'Sleep Patterns']

for col in binary_columns:
    df[col] = label_encoder.fit_transform(df[col])

# 2. One-Hot Encoding for multi-class categorical features
multi_class_columns = ['Previous Complications', 'Pregnancy Type', 'Chronic Conditions', 'Allergies',
                       'Medications', 'Family History', 'Physical Activity Level', 'Stress Level',
                       'Placenta Condition', 'Amniotic Fluid Levels', 'Preferred Notification Method', 'Fatigue Level',
                       'Frequency of Notifications', 'Feedback on Symptoms' ]

# Apply one-hot encoding to the multi-class columns
df = pd.get_dummies(df, columns=multi_class_columns, drop_first=True)


```

```python
import pandas as pd
import numpy as np

# Assuming `df` is your original DataFrame
# Copy the data for augmenting
augmented_data = df.copy()

# Randomly add noise to the numerical features
def add_noise_to_numerical_features(df):
    numerical_columns = df.select_dtypes(include=[np.number]).columns
    noise_factor = 0.01  # Add 1% noise to numerical values

    # Add noise to numerical columns
    for col in numerical_columns:
        df[col] = df[col] + np.random.normal(0, noise_factor * df[col].std(), size=df.shape[0])

    return df

# Add noise to the numerical features of the dataset
augmented_data = add_noise_to_numerical_features(augmented_data)

# Concatenate the original dataset with the augmented dataset
df_doubled = pd.concat([df, augmented_data], ignore_index=True)

# Check the resulting shape to ensure the data is doubled
print(f"Original Data Shape: {df.shape}")
print(f"Doubled Data Shape: {df_doubled.shape}")
```

```python
import pandas as pd
from datetime import datetime

# Assuming 'LMP' and 'EDD' are in 'YYYY-MM-DD' format as strings
# Convert to datetime
df['Last Menstrual Period (LMP)'] = pd.to_datetime(df['Last Menstrual Period (LMP)'], errors='coerce')
df['Estimated Due Date (EDD)'] = pd.to_datetime(df['Estimated Due Date (EDD)'], errors='coerce')

# Calculate Weeks of Pregnancy
df['Pregnancy Start Date'] = df['Last Menstrual Period (LMP)']  # The date the pregnancy started
df['Current Date'] = datetime.now()  # The current date

# Calculate the difference in days between current date and LMP
df['Days Since LMP'] = (df['Current Date'] - df['Pregnancy Start Date']).dt.days

# Convert Days to Weeks for easier interpretation
df['Weeks of Pregnancy'] = df['Days Since LMP'] // 7

# Calculate the difference between the Estimated Due Date and Current Date to get days until delivery
df['Days Until Delivery'] = (df['Estimated Due Date (EDD)'] - df['Current Date']).dt.days

# Convert Days Until Delivery to Weeks
df['Weeks Until Delivery'] = df['Days Until Delivery'] // 7

# Calculate Total Pregnancy Duration
df['Total Pregnancy Duration (Days)'] = (df['Estimated Due Date (EDD)'] - df['Last Menstrual Period (LMP)']).dt.days

# Drop the original LMP and EDD columns as we no longer need them
df = df.drop(columns=['Last Menstrual Period (LMP)', 'Estimated Due Date (EDD)'])# Drop the original date columns as they are no longer needed


# Check the updated data
# print(df[['Days Since LMP', 'Weeks of Pregnancy', 'Days Until Delivery', 'Weeks Until Delivery', 'Total Pregnancy Duration (Days)']].head())
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, mean_absolute_error, mean_squared_error, r2_score
from xgboost import XGBClassifier, XGBRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.multioutput import MultiOutputClassifier, MultiOutputRegressor
```

```python
# Assuming 'df' is your dataframe
X = df.drop(columns=['Risk Assessment', 'Water Breakage Likelihood', 'Days Until Delivery', 'Weeks Until Delivery'])

# Target columns for classification and regression
y_classification = df[['Risk Assessment', 'Water Breakage Likelihood']]
y_regression = df[['Days Until Delivery', 'Weeks Until Delivery']]

print(X.columns)

# Scaling the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

```

```python
example_row = X.iloc[0]  # Get the first row
for column in X.columns:
    print(f"{column}: {example_row[column]}")
```

```python
# Split the data into training and testing sets for both classification and regression
X_train, X_test, y_train_class, y_test_class = train_test_split(X_scaled, y_classification, test_size=0.2, random_state=42)
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_scaled, y_regression, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
le = LabelEncoder()

# Encode the target variables for classification
y_train_class_encoded = y_train_class.apply(le.fit_transform)
y_test_class_encoded = y_test_class.apply(le.transform)
```

```python
from sklearn.multioutput import MultiOutputClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Initialize the XGBoost Classifier
xgb_class = XGBClassifier(eval_metric='mlogloss', use_label_encoder=False)

# Wrap the XGBoost Classifier inside MultiOutputClassifier
multi_target_model = MultiOutputClassifier(xgb_class, n_jobs=-1)

# Fit the model on the training data
multi_target_model.fit(X_train, y_train_class_encoded)

# Make predictions
y_pred_class = multi_target_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Evaluate the model for Risk Assessment (first target)
print("\nClassification Report for Risk Assessment:")
print(classification_report(y_test_class_encoded['Risk Assessment'], y_pred_class[:, 0]))  # First column for Risk Assessment

print("\nConfusion Matrix for Risk Assessment:")
print(confusion_matrix(y_test_class_encoded['Risk Assessment'], y_pred_class[:, 0]))  # First column for Risk Assessment

# Evaluate the model for Water Breakage Likelihood (second target)
print("\nClassification Report for Water Breakage Likelihood:")
print(classification_report(y_test_class_encoded['Water Breakage Likelihood'], y_pred_class[:, 1]))  # Second column for Water Breakage Likelihood

print("\nConfusion Matrix for Water Breakage Likelihood:")
print(confusion_matrix(y_test_class_encoded['Water Breakage Likelihood'], y_pred_class[:, 1]))  # Second column for Water Breakage Likelihood
```

```python
from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Initialize Gradient Boosting Regressor
gbr = GradientBoostingRegressor(n_estimators=100, random_state=42)

# Wrap the model with MultiOutputRegressor to handle multiple outputs
multi_target_model = MultiOutputRegressor(gbr)

# Fit the model on the training data
multi_target_model.fit(X_train_reg, y_train_reg)

# Make predictions
y_pred_reg = multi_target_model.predict(X_test_reg)


```

```python
# Evaluate the model for regression
print("\nRegression Metrics for Days Until Delivery and Weeks Until Delivery:")
print(f"Mean Absolute Error (MAE): {mean_absolute_error(y_test_reg, y_pred_reg)}")
print(f"Mean Squared Error (MSE): {mean_squared_error(y_test_reg, y_pred_reg)}")
print(f"R² Score: {r2_score(y_test_reg, y_pred_reg)}")
```
